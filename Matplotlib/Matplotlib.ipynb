```python
import matplotlib.pyplot as plt

#Agar dapat menampilkan plot di jupyter
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,31)
y = x**2
print(x)
print()
print(y)
```

## Functional

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My First Plot')



plt.show()
```

 > **Subplot**

```python
#n-row, n-column, index plot (urutan plot)
#1/1. 1/2, 2
plt.subplot(1, 2, 2)
plt.plot(x, y, 'r')

#n-row, n-column, index plot (urutan plot)
#1/1. 1/2, 2
plt.subplot(1, 2, 1)
plt.plot(y, x, 'g')


plt.show()
```

```python
#Membuat kanvas kosong
fig = plt.figure()

#Menambah axes
#Left, Bottom, Width, & height
axes = fig.add_axes([0, 0, 0.9, 0.9])

#set plot
axes.plot(x,y)

#set label
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('My Plot')

plt.show()
```

```python
fig = plt.figure()

#Menambah axes
#Left, Bottom, Width, & height
axes1 = fig.add_axes([0, 0.6, 0.9, 0.4])
axes2 = fig.add_axes([0, 0, 0.9, 0.4])

#set title
axes1.set_title('Axes One')
axes2.set_title('Axes Two')

#set plot
axes1.plot(x,y,'r')
axes2.plot(y,x)
```
